```python
def quick_sort(data, column_index):
    if len(data) <= 1:
        return data
    else:
        pivot = data[0][column_index]
        left = []
        right = []
        for i in range(1, len(data)):
            if data[i][column_index] < pivot:
                left.append(data[i])
            else:
                right.append(data[i])
        return quick_sort(left, column_index) + [data[0]] + quick_sort(right, column_index)

def index_sort(lst):
    indexed_list = []
    for index, element in enumerate(lst):
        indexed_list.append((index, element))
    # print(indexed_list)

    sorted_list = quick_sort(indexed_list, 1)
    result = []
    for i in sorted_list:
        result.append(i[0])
    return result
```

```python
def mode(lst):
    temp = {}
    for i in lst:
        if i not in temp:
            temp[i] = 1
        else:
            count = temp[i] + 1
            temp[i] = count
    maxim = 0
    label_maxim = []
    for i in temp:
        if temp[i] > maxim:
            maxim = temp[i]
            label_maxim = i
    return label_maxim
```

```python
def measures(k, test_2D, dists):
    row = 0
    for lst in dists:
        sorted_list = index_sort(lst)
        knn_result = sorted_list[:k]
        
        label_knn = []
        for j in knn_result:
            label = trained['Sentiment'][j]
            label_knn.append(label)
        # print(label_knn)
        
        max_label = mode(label_knn)
        # print(max_label)
        # test.to_csv('Tweets_Test.csv', index=False)
        # test['Predicted'][row] = max_label
        test.loc[row, 'Predicted'] = max_label
        # trained.loc[row, 'Predicted'] = max_label
        # trained.to_csv('./Tweets.csv', index=False)
        test.to_csv('./Tweets_Test.csv', index=False)
        row += 1

        
measures(1, test_2D, dists)
```
